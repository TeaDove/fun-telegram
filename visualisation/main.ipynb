```python
def draw_values(df: pd.DataFrame) -> None:
    ax = plt.gca()
    
    for color in df['color'].unique():
        colored_df = df[df['color'] == color] 
        colored_df.plot(kind='scatter', x='lon', y='lat',grid=True, stacked=True, ax=ax, color=color, s=10)
    plt.show()

# draw_values(points)
```
